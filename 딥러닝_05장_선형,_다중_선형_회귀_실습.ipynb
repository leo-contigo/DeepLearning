```python
import numpy as np
import matplotlib.pyplot as plt

# 그래프 설정 
plt.rcParams['font.family'] = 'malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 13
plt.rcParams['figure.figsize'] = 15, 10
plt.rcParams['axes.unicode_minus'] = False

# 그래프 설정 (코랩 한글)
plt.rc('font' , family='NanumBarunGothic')
```

```python
# 공부 시간 x, 성적 y, 넘파이 배열

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# 데이터의 분포를 그래프로 
plt.scatter(x, y)
plt.show()
```

```python
# 기울기 a값과 절편 b의 값을 초기화
a = 0
b = 0

# 학습률
lr = 0.03

# 몇번 반복할지
epochs = 2001

# x값이 총 몇개인지
n = len(x)

# 경사하강법

# epochs만금 반복
for i in range(epochs):
  # 예측값 구하는 공식
  y_pred = a * x + b

  # 실제값과 비교한 오차를 error
  error = y - y_pred

  # 오차함수를 a로 편미분한 값
  a_diff = (2/n) * sum(-x * (error))
   # 오차함수를 b로 편미분한 값 
  b_diff = (2/n) * sum(-(error))

  # 학습률을 곱해 기존이 a값을 업데이트
  a = a-lr * a_diff
  # 학습률을 곱해 기존이 b값을 업데이트
  b = b-lr * b_diff

  # 100반 반복될 때 현재의 a, b값을 출력
  if i % 100 == 0:
    print('epochs = %.f, 기울기 = %.04f, 절편 = %.04f' % (i, a, b))

# 최종 a값을 기울기 b값을 y절편에 대입해 시각화
y_pred = a * x + b

plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()
```

# 다중 선형 회귀

```python
# x 값 하나 더 추가

# 공부시간 x1, 과외시간 x2, 성적 y의 데이터 값
x1 = np.array([2,4,6,8])
x2 = np.array([0, 4, 2, 3])

y = ([81, 93, 91, 97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y);
plt.show()
```

```python
# x값이 두 개가 되었으니, 공부시간 x1, 과외시간 x2
# 기울기도 a1, a2 이렇게 두 개를 만듬. 
a1 = 0
a2 = 0
b = 0

# 학습률 정하기
lr = 0.01
# 반복
eppchs = 2001

n = len(x1)

# 경사 하강법
for i in range(epochs):
  # 기울기와 절편 자리에 a1, a2, b를 각각 대입
  y_pred = a1 * x1 + a2 * x2 + b

  # 실제 값과 비교한 오차를 error
  error = y - y_pred

  # 오차함수 a1, a2, b를 편미분한 값
  a1_diff = (2/n) * sum (-x1 * (error))
  a2_diff = (2/n) * sum(-x2 * (error))
  b_diff = (2/n) * sum(-(error))

  # 학습률을 곱해 기존의 기울기와 절편을 업데이트
  a1 = a1 - lr * a1_diff
  a2 = a2 - lr * a2_diff
  b = b - lr * b_diff

  # 100반 반복될 때 현재의 a1, a2, b값을 출력
  if i % 100 == 0:
    print('epochs = %.f, 기울기1 = %.04f, 기울기2 = %0.4f, 절편 = %.04f' % (i, a1, a2, b)) 
```

선형 회귀 : 현상을 분석하는 방법
머신러닝은 이러한 분석방법을 통해 예측 모델을 만드는 것.

|머신러닝|딥러닝|
|----|--------|
|y=ax + b|H(x) = wx + b / w: 가중치, b: 편향(bias)|
|평균제곱오차|손실함수(loss function)|
|경사하강법|옵티마이저|

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential()

# h(x) = wx + b
# 출력값 = 성적 : 하나씩이므로 Dense()함수 첫번째는 1, 
# 입력될 변수 = 학습시간 : 하나뿐이므로 input_dim = 1
# 입력된 값을 다음 층으로 넘길 때 각 값을 어떻게 처리할지 결정하는 함수 = 활성화 함수
# activation = 활성화 함수
# 선형회귀 : linear
model.add(Dense(1, input_dim=1, activation='linear'))

# 옵티마이저 sgd라고 설정 
# 손실함수는 평균제곱오차 사용(mse)
model.compile(optimizer='sgd', loss='mse')
model.fit(x, y, epochs=2000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

# 임의 시간
hour = 7
prediction = model.predict([hour])
print('%.f시간을 공부할 경우 예상 점수는 %.2f점입니다.' %(hour, prediction))
```

변수가 두개인 다중선형회귀

```python
# 변수가 2개 input_dim = 2

x1 = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81,93,91,97])

model = Sequential()

# h(x) = wx + b
# 출력값 = 성적 : 하나씩이므로 Dense()함수 첫번째는 1, 
# 입력될 변수 = 학습시간 : 하나뿐이므로 input_dim = 1
# 입력된 값을 다음 층으로 넘길 때 각 값을 어떻게 처리할지 결정하는 함수 = 활성화 함수
# activation = 활성화 함수
# 선형회귀 : linear
model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd' ,loss='mse')

model.fit(x1, y, epochs=2000)

# 임의 시간: 학습시간, 과외시간
hour = 7
private_class = 4
prediction = model.predict([[hour, private_class]])
print('%.f시간을 공부하고 %.f시간 과외할 경우 예상 점수는 %.2f점입니다.' %(hour, private_class, prediction))
```
