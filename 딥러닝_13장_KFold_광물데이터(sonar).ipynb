````python
df = pd.read_csv('./data/sonar3.csv', header = None)
df.head()

# 0번부터 59번까지 음파 주파수의 에너지를 0에서 1사이의 숫자로 표현
# 60번 : 일반 암석 0, 광석 1
````

````python
# 정보와 결측치 확인
df.info()
df.isna().sum()
````

````python
# 결과 데이터 분리하기
X = df.drop([60], axis=1)
y = df[60]

display(X)
display(y)
````

````python
# 신경망 모델을 관리하는 객체
from tensorflow.keras.models import Sequential
# 선형 회귀 레이어
from tensorflow.keras.layers import Dense
# 활성화 함수를 정의하는 객체
from tensorflow.keras.layers import Activation

# 모델 설정
# 신경망 설계 
model = Sequential()
# 은닉층
model.add(Dense (24, input_dim = 60))
# 활성화 함수
model.add(Activation('relu'))

model.add(Dense(10, activation = 'relu'))

# 출력층
model.add(Dense(1, activation = 'sigmoid'))

# 모델 컴파일
# 2진분류! : binary_crossentropy
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
model.summary()
````

````python
# 모델 실행
model.fit(X, y, epochs=200, batch_size=10)
````

````python
# 정확도 확인
a = model.evaluate(X, y)
a
print(f'손실률 : {a[0]}')
print(f'정확도 : {a[1]}')
````

````python
# 신경망 모델을 관리하는 객체
from tensorflow.keras.models import Sequential
# 선형 회귀 레이어
from tensorflow.keras.layers import Dense
# 활성화 함수를 정의하는 객체
from tensorflow.keras.layers import Activation

# 모델 설정
# 신경망 설계 
model = Sequential()
# 은닉층
model.add(Dense (24, input_dim = 60))
# 활성화 함수
model.add(Activation('relu'))

model.add(Dense(10, activation = 'relu'))

# 출력층
model.add(Dense(1, activation = 'sigmoid'))

# 모델 컴파일
# 2진분류! : binary_crossentropy
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
model.summary()
````

````python
# 모델 실행
model.fit(X_train, y_train, epochs=200, batch_size=10)
````

````python
# 정확도 확인
b = model.evaluate(X_test, y_test)
print(f'손실률 : {b[0]}')
print(f'정확도 : {b[1]}')
````

전체 데이터에서 7:3으로 설정할 경우 30%의 테스트셋은 학습에 사용할 수 없다는 단점.<br>

이를 해결하고자 
```k겹 교차검증(K-fold cross calidation)```<br> 
K겹 교차 검증이란?<br>
    - 데이터셋을 여러 개로 나누어 하나씩 테스트셋으로 사용하고 나머지를 모두 합해서 학습셋으로 사용하는 방법.<br>
    이렇게 하면 데이터 100%를 학습데이터로 사용할 수 있고, 테스트데이터로도 사용할 수 있음.

````python
# 사이킷런 KFold() 함수 
from sklearn.model_selection import KFold

from sklearn.metrics import accuracy_score

# 몇개의 파일로 나눌것인가 ? 5
k = 5

#KFold 함수를 불러옵니다. 분할하기 전에 샘플이 치우치지 않도록 섞어 줍니다.
kfold = KFold(n_splits=k, shuffle=True)

# 정확도 채울 리스트
acc_score = []

def model_fn():
    model = Sequential()
    model.add(Dense(24, input_dim = 60, activation = 'relu'))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))

    return model

# k개의 학습데이터, 테스트데이터로 분리되면서 반복
for train_index , test_index in kfold.split(X): 
    X_train , X_test = X.iloc[train_index,:], X.iloc[test_index,:]  
    y_train , y_test = y.iloc[train_index], y.iloc[test_index]

    model = model_fn()
    model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    history = model.fit(X_train, y_train, epochs = 200, batch_size=10, verbose=0)

    # 정확도 
    accuracy = model.evaluate(X_test, y_test)[1]
    # 리스트에 저장
    acc_score.append(accuracy)

avg_score = sum(acc_score) / k

# 제발ㅋㅋㅋ 철자좀ㅋㅋㅋㅋ 틀리지 말자
````

````python
print(f'정확도 : {acc_score}')
print(f'정확도 평균 : {avg_score}')

# 5번의 정확도
````
